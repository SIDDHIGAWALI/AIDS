```python
object_columns = df.select_dtypes(include=['object']).columns
print("Object Type Columns : ")
print(object_columns)

numerical_columns = df.select_dtypes(include=['int64','float64']).columns
print("\nNumerical Type Columns : ")
print(numerical_columns)
# Finding Obejct column and numerical column
```

```python
def classify_features(df):
    categorical_features = []
    non_categorical_features = []
    discrete_features = []
    continuous_features = []

    for column in df.columns:
        if df[column].dtype == 'object':
            if df[column].nunique() < 10:
                categorical_features.append(column)
            else:
                non_categorical_features.append(column)
        elif df[column].dtype in ['int64','float64']:
            if df[column].nunique() < 10:
                discrete_features.append(column)
            else:
                continuous_features.append(column)
    return categorical_features, non_categorical_features, discrete_features, continuous_features
```

```python
categorical, non_categorical, discrete, continuous = classify_features(df)
print("Categorical Features :", categorical)
print("***************")
print("Non-Categorical Features :", non_categorical)
print("***************")
print("Discrete Features :", discrete)
print("***************")
print("Continuous Features :", continuous)
print("***************")
```

```python
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix.
correlation_matrix = numerical_df.corr()

# Display the correlation matrix.
print(correlation_matrix)
```

```python
# Scatter Plot: Rent Index vs Cost of Living Index
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Rent_Index", y="Cost_of_Living_Index", hue="Country", legend=False)
plt.title("Scatter Plot: Rent Index vs Cost of Living Index")
plt.xlabel("Rent Index")
plt.ylabel("Cost of Living Index")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[["Cost_of_Living_Index", "Rent_Index", "Cost_of_Living_Plus_Rent_Index",
                           "Groceries_Index", "Restaurant_Price_Index", "Local_Purchasing_Power_Index"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Cost_of_Living_Index", color="blue")
plt.title("Histogram: Distribution of Cost of Living Index")
plt.xlabel("Cost of Living Index")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y="Cost_of_Living_Index", color="purple")
plt.title("Box Plot: Cost of Living Index")
plt.ylabel("Cost of Living Index")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Country", y="Rent_Index")
plt.title("Bar Plot: Rent Index (Top 10 Countries)")
plt.xlabel("Country")
plt.ylabel("Rent Index")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df.nsmallest(10, 'Rank')
plt.figure(figsize=(10, 6))
plt.plot(data["Country"], data["Cost_of_Living_Index"], color="green")
plt.title("Line Graph: Cost of Living Index (Top 10 Countries)")
plt.xlabel("Country")
plt.ylabel("Cost of Living Index")
plt.show()
```

```python
data = df.nlargest(5, "Cost_of_Living_Index")
plt.figure(figsize=(8, 8))
plt.pie(data["Cost_of_Living_Index"], labels=data["Country"], colors=sns.color_palette("pastel"))
plt.title("Pie Chart: Cost of Living Index (Top 5 Countries)")
plt.show()
```
